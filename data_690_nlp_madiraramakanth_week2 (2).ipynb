```python
import requests
from bs4 import BeautifulSoup
url = requests.get('https://pythonprogramming.net/parsememcparseface/')
print(url)
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
print("Page Title:", soup.title.text)
```

```python
attributes = soup.find_all()
for attr in attributes:
    print("Tag:", attr.name)
    print("Attributes:", attr.attrs)
```

```python
all_paragraphs = soup.find_all('p')

# 7. Iterate through paragraphs and print text
for i in all_paragraphs:
    print("Paragraph Text:", i.text)
```

```python
get_links = soup.find_all('a')
for i in get_links:
    print("Link:", i.get('href'))
```

```python
only_text = soup.get_text()
print("Text Only:", only_text)
```
